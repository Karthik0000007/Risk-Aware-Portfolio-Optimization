```python
import yfinance as yf

etfs = ['SPY', 'QQQ', 'IWM', 'TLT', 'GLD', 'XLF', 'XLK', 'XLE']

etf_data = yf.download(etfs, start="2019-01-01", end="2024-12-31", auto_adjust=False)['Adj Close']
etf_data.to_csv("../data/etf_data.csv")

etf_data.plot(figsize=(14, 6), title="ETF Prices")

vix_data = yf.download('^VIX', start="2019-01-01", end="2024-12-31", auto_adjust=False)['Adj Close']

vix_data.to_csv("../data/vix_data.csv")

vix_data.plot(figsize=(12, 4), title="VIX - Volatility Index")
```

```python
vix_data = yf.download('^VIX', start="2019-01-01", end="2024-12-31", auto_adjust=False)

# Use the 'Close' column (most consistent for indices like VIX)
vix_data = vix_data[['Close']].rename(columns={'Close': 'VIX'})

# Save to CSV
vix_data.to_csv("../data/vix_data.csv")

# Plot
vix_data.plot(figsize=(12, 4), title="VIX - Volatility Index")
plt.show()
```
